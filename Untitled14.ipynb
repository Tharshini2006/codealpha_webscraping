```python
import requests
from bs4 import BeautifulSoup
```

```python
web=requests.get("https://www.geeksforgeeks.org/dsa/dsa-sheet-by-love-babbar/")
```

```python
web
```

```python
soup = BeautifulSoup(web.content,"html.parser")
print(soup.prettify)
```

```python
for i in soup.find_all("a"):
  print(i.get("href"))
```

```python
img = soup.find_all("img")
img
```

```python
for i in img:
  print(i.get("src"))
```
